```python
#the details to be put on the list
more_details=[]

for a2 in content:
    temp2= a2.find_all("span", class_="sc-be6f1408-8 fcCUPU cli-title-metadata-item") #find_all for all the types in the same class 
    for a3 in temp2:
     more_details.append(a3.text)
     
print(more_details)
len(more_details)
```

```python
#Filtering the data out of the list and appending it to its appropriate list

years = []
times = []
ratings = []

for i in range(0, len(more_details)-1, 3):
    years.append(more_details[i])
    
    times.append(more_details[i+1])

    ratings.append(more_details[i+2])

result = list(zip(years, times, ratings))
print(result)
print("Year of release-",years)
print("Length of the movie-",times)
print("Age Rating of the movie-",ratings)


#the result is the list elements divided into 3 elements and all appended to different appropriate list
```

```python
#Now to divide it more into list

rating=[]
rated=[]

for i in split:
  rating.append(i[0])
  rated.append(i[1])
    
print(rating)
print(rated)

#Use for loop if only 2 values in list. Use range if more

#this works too
#for i in range(0, len(split), 2):
   # rating.append(split[i][0])
    #rated.append(split[i][1])
    
```

```python
#splitting into two elements in a list

t= [i.split(".") for i in titles]
t

#creaing a list to be appended to
final_title=[]

#for loop to append the [1] elements from the list t
for temp in t:
  final_title.append(temp[1])
  
#stripping the extra blank space
final_title = [f.strip(" ") for f in final_title]

final_title
```

```python
#appending all list in one
final= list(zip(final_title,years, times, ratings,rating,rated))
final
```

```python
#reating a data frame out of it
df= pd.DataFrame(final,columns=['title','year', 'times', 'maturity','rating','rated'])
```
